Load data, define functions and plot figure. 

```python
# read in bar chart data
outdir = '/g/data/e14/hd4873/runs/parcels/output/AntConn/data/2Dpxy/'
outfile = outdir + 'BarChartData_AS-BS.nc'
data = xr.open_dataset(outfile)
```

```python
fig = plt.figure(constrained_layout=True,figsize=(12,5))
spec1 = gridspec.GridSpec(ncols=2, nrows=1, figure=fig)
ax1 = fig.add_subplot(spec1[0, 0], )
ax2 = fig.add_subplot(spec1[0, 1], )
regions = ['EA-W', 'EA-E', 'RS', 'WA', 'WAP', 'WS']
c = 'midnightblue'

barlist = ax1.bar(data.downstream_region, data.basin_trans_pct[0,:], color=c, alpha=0.9, zorder=2)
barlist[3].set_color('none')
ax1.bar(data.downstream_region[3], data.basin_trans_pct[0,3], edgecolor=c, color='w', zorder=2)
#ax1.bar(data.downstream_region[3], data.basin_trans_pctWA[0,3], bottom =data.basin_trans_pctWA[0,3], edgecolor=c, color='w', zorder=2)
ax1.bar(data.downstream_region[3], data.basin_trans_pctWA[0,3], edgecolor=c, color='lightsteelblue', zorder=2, alpha=0.6)
ax1.grid(c='gainsboro')
ax1.set_ylabel('Connectivity (%)', fontsize=15)
ax1.tick_params(axis='y',labelsize=15)
ax1.tick_params(axis='x',labelsize=15)
ax1.set_xticklabels(labels = regions, rotation=90, ha='center', va='top')

barlist = ax2.bar(data.downstream_region, data.basin_trans_pct[1,:], color=c, alpha=0.9, zorder=2)
barlist[3].set_color('none')
ax2.bar(data.downstream_region[3], data.basin_trans_pct[0,3], edgecolor=c, color='w', zorder=2)
ax2.bar(data.downstream_region[3], data.basin_trans_pctWA[1,3], bottom =data.basin_trans_pctWA[0,3], edgecolor=c, color='lightsteelblue', alpha=0.6, zorder=2)
ax2.bar(data.downstream_region[3], data.basin_trans_pctWA[0,3], edgecolor=c, color='w', zorder=2)
ax2.grid(c='gainsboro')
#ax2.set_ylabel('Transport (%)', fontsize=15)
ax2.tick_params(axis='y',labelsize=15)
ax2.tick_params(axis='x',labelsize=15)
ax2.set_xticklabels(labels = regions, rotation=90, ha='center', va='top')

ax1.text(-0.5, 97, 'a)', color='k', fontsize=22,fontweight='bold')
ax2.text(-0.5, 97, 'b)', color='k', fontsize=22,fontweight='bold')

ax1.text(4, 86, 'Amundsen\n Sea', color='dimgrey', fontsize=15,
        bbox=dict(facecolor='none', edgecolor='dimgrey', boxstyle='round,pad=0.3'))
ax2.text(3.6, 86, 'Bellingshausen\n Sea', color='dimgrey', fontsize=15,
        bbox=dict(facecolor='none', edgecolor='dimgrey', boxstyle='round,pad=0.3'))

ax1.text(0.08, 0.18, '17', fontweight='normal',c=c, fontsize=15, transform=ax1.transAxes)
ax1.text(0.243, 0.44, '44', fontweight='normal',c=c, fontsize=15, transform=ax1.transAxes)
ax1.text(0.4, 0.89, '92', fontweight='normal', c=c, fontsize=15, transform=ax1.transAxes)
ax1.text(0.723, 0.06, '4', fontweight='normal',c=c, fontsize=15, transform=ax1.transAxes)
ax1.text(0.882, 0.06, '5', fontweight='normal',c=c, fontsize=15, transform=ax1.transAxes)
ax1.text(0.56, 0.26, '52', fontweight='normal',c=c, fontsize=15, transform=ax1.transAxes)

ax2.text(0.1, 0.1, '8', fontweight='normal',c=c, fontsize=15, transform=ax2.transAxes)
ax2.text(0.245, 0.21, '20', fontweight='normal',c=c, fontsize=15, transform=ax2.transAxes)
ax2.text(0.4, 0.42, '42', fontweight='normal', c=c, fontsize=15, transform=ax2.transAxes)
ax2.text(0.715, 0.21, '20', fontweight='normal',c=c, fontsize=15, transform=ax2.transAxes)
ax2.text(0.882, 0.05, '3', fontweight='normal',c=c, fontsize=15, transform=ax2.transAxes)
ax2.text(0.56, 0.68, '48', fontweight='normal',c=c, fontsize=15, transform=ax2.transAxes)

# save to file
outdir = '/g/data/e14/hd4873/runs/parcels/output/AntConn/figures/for_publication/'
outfile = outdir + 'FigS4_AS-BS_connectivity_bar_chart.jpg'
fig.savefig(outfile, dpi=300, bbox_inches='tight')
```
